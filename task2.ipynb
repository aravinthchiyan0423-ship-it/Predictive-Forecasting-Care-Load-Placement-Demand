```python
from sklearn.metrics import mean_absolute_error
import numpy as np

def walk_forward_validate(model_fit_func, df, target_col, window=30, horizon=7):
    errors = []
    for i in range(window, len(df) - horizon, horizon):
        train_chunk = df.iloc[:i]
        test_chunk = df.iloc[i:i + horizon]
        preds = model_fit_func(train_chunk, horizon)
        mae = mean_absolute_error(test_chunk[target_col], preds)
        errors.append(mae)
    return np.mean(errors)
```
